```python
import pandas as pd
df = pd.read_excel('Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.xlsx')
df
```

```python
#compile reviews
text = df['reviews.text']
text2 = df['reviews.title']
final_text = str()
for t in text:
  final_text += str(t)
for t in text2:
  final_text += str(t)  
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 
# stopwords is a collection of words that dont convey meaning. mostly pronouns such as he she etc.

wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='white', collocations=False, stopwords = STOPWORDS).generate(final_text)
plt.figure(figsize=(30, 20))
plt.imshow(wordcloud) 
plt.axis("off")
plt.show()
```
